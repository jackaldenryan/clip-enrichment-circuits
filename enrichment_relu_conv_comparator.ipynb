```python
from PIL import Image
from activations_tool import activations, get_layer
import torch
```

```python
import scipy.spatial

def get_percent_residual_contribution(unit: int, image: Image):
    residual_act = activations(get_layer(4, 0, 7), image)[0][unit]
    conv_act = activations(get_layer(4, 0, 5), image)[0][unit]

    magnitude = torch.linalg.norm(residual_act) / (torch.linalg.norm(residual_act) + torch.linalg.norm(conv_act))
    similarity = scipy.spatial.distance.cosine(residual_act.cpu().flatten(), conv_act.cpu().flatten())

    return (magnitude, similarity)
```

```python
for unit in range(2048):
    image = Image.open(f"../../seri/datasets/scrape4_0_8/channel{unit}.png")
    image = image.convert("RGB")
    print("Unit", unit, "->", get_percent_residual_contribution(unit, image))
```
