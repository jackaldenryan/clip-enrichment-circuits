# 100k word baseline calculation

```python
layer = activations_tool.get_layer(4,2,6)

baseline = np.zeros((2048))

for idx, line in enumerate(open("datasets/google-10000-english-usa.txt")):
    word = line.strip()
    
    _, _, activations = activations_tool.highest_channel_for_image(layer, rasterize_text(word))
    
    baseline = baseline + activations
    
    if idx % 10 == 0:
        print(idx, "norm:", np.linalg.norm(baseline))
```

```python
from scipy import spatial

layer = activations_tool.get_layer(4,2,6)

_, _, eee_acts = activations_tool.highest_channel_for_image(layer, rasterize_text("eeeeee"))

print(np.linalg.norm(eee_acts))
spatial.distance.cosine(baseline, eee_acts)
```

# Using the Channel Difference Engine

```python
import matplotlib.pyplot as plt
import numpy as np

target_text = "Trump"
baseline_text = "eeeeee"
layer = activations_tool.get_layer(3,5,6)

print("Inspecting layer type:", layer)

print("Baseline activation histogram:")
baseline_image = rasterize_text(baseline_text)
_, _, baseline_acts = activations_tool.highest_channel_for_image(layer, baseline_image)
plt.hist(baseline_acts, bins=100)
plt.show()

print("Rasterized image:")
image = rasterize_text(target_text)
display(image)

activation_differences = activations_tool.channel_difference_engine(layer, image, baseline_image)
unit, max_act, all_acts = activations_tool.highest_channel_for_image(layer, image)

acts = np.subtract(all_acts, baseline_acts)
i = np.argmax(acts)
print("Maximally activating channel after baseline subtraction:", i)

print("Top ten activating channels:")
top_ten = list(activation_differences.items())[:10]
for channel, act in top_ten:
    print(f"{channel}: activation {act}")

print("Histogram of activations - baseline:")
plt.hist(acts, bins=100)
plt.show()
print('Maximally activating channel + activation without baseline subtraction:', unit, max_act)
```

```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np

filename = "/usr/share/fonts/truetype/noto/NotoSans-Regular.ttf"
DEFAULT_FONT = ImageFont.truetype(filename, 40)

def draw_text(text_str, font, should_show=True):
    blank = np.ones((224, 224))
    blank = np.uint8(blank) * 255
    im = Image.fromarray(blank)
    if im.mode != "RGB":
        im = im.convert("RGB")
    draw = ImageDraw.Draw(im)
    draw.text(
        (20 + 50 * 0, 20 + 50 * 2), text_str, (0, 0, 0), font=font,
    )
    draw = ImageDraw.Draw(im)
    if should_show:
        display(im)
    return np.asarray(im) / 255

draw_text("chicken", DEFAULT_FONT)
```
